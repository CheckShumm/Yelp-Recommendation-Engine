```python
# Function to calculate distance from a restaurant
def getDistance(lat, lon):
    # approximate radius of earth in km
    R = 6373.0
    myLat = 45.393220
    myLon = -73.493000

    dlon = myLon - lon
    dlat = myLat - lat

    a = np.sin(dlat / 2)**2 + np.cos(lat) * np.cos(myLat) * np.sin(dlon / 2)**2
    c = 2 * np.arctan2(np.sqrt(a), np.sqrt(1 - a))

    distance = R * c
    
    return distance 
```
